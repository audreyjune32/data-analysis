```python
#5 Charts
###Line chart
# List of countries
countries = ['Kenya', 'United States', 'India']

# Set figure size
plt.figure(figsize=(12, 6))

# Loop through countries and plot
for country in countries:
    country_data = df[df['location'] == country]
    plt.plot(
        country_data['date'],
        country_data['total_vaccinations'],
        label=country
    )

# Chart settings
plt.title("Cumulative COVID-19 Vaccinations Over Time", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Total Vaccinations", fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
##5Charts
## pie chart
# Select latest data for India
india = df[(df['location'] == 'India') & (df['date'] == df['date'].max())]

# Get values
vaccinated = india['people_fully_vaccinated_per_hundred'].values[0]
unvaccinated = 100 - vaccinated

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie([vaccinated, unvaccinated],
        labels=['Vaccinated', 'Unvaccinated'],
        autopct='%1.1f%%',
        colors=['green', 'lightgray'],
        startangle=140)
plt.title("India: Vaccinated vs Unvaccinated")
plt.axis('equal')
plt.show()
```

```python
##6. Choropleth Map with Plotly Express
#a. Prepare the dataframe


# Ensure 'date' column is datetime
df['date'] = pd.to_datetime(df['date'])

# Get the most recent date per country
latest_df = df.sort_values('date').groupby('location', as_index=False).last()

# Keep only needed columns and drop missing data
choropleth_df = latest_df[['location', 'iso_code', 'total_cases', 'people_fully_vaccinated_per_hundred']].dropna(subset=['iso_code', 'total_cases'])
choropleth_df 
```

```python
##6b ..Plot Choropleth for Total cases
import plotly.express as px

fig = px.choropleth(
    choropleth_df,
    locations='iso_code',
    color='total_cases',
    hover_name='location',
    color_continuous_scale='Reds',
    title='Total COVID-19 Cases by Country (Latest Available Data)'
)

fig.update_geos(showcountries=True, showcoastlines=True)
fig.show()
```

```python
fig = px.choropleth(
    choropleth_df,
    locations='iso_code',
    color='people_fully_vaccinated_per_hundred',
    hover_name='location',
    color_continuous_scale='Blues',
    title='🌍 % of Population Fully Vaccinated by Country'
)

fig.update_geos(showcountries=True, showcoastlines=True)
fig.show()
```

```python
###4
##a) plot cases over time


countries = ['Kenya', 'United States', 'India']
plt.figure(figsize=(12, 6))

for country in countries:
    data = df[df['location'] == country]
    plt.plot(data['date'], data['total_cases'], label=country)

plt.title("Total COVID-19 Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
##4b
##plot total deaths over time
plt.figure(figsize=(12, 6))

for country in countries:
    data = df[df['location'] == country]
    plt.plot(data['date'], data['total_deaths'], label=country)

plt.title("Total COVID-19 Deaths Over Time")
plt.xlabel("Date")
plt.ylabel("Total Deaths")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
##4c) compare daily new cases between countries
plt.figure(figsize=(12, 6))

for country in countries:
    data = df[df['location'] == country]
    plt.plot(data['date'], data['new_cases'], label=country)

plt.title("Daily New COVID-19 Cases")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
##4d) Calculate and display death raate (as %)

# Create a new column for death rate
df['death_rate'] = df['total_deaths'] / df['total_cases'] * 100

# Show latest death rate for each country
latest = df.sort_values('date').groupby('location', as_index=False).last()

for country in countries:
    rate = latest[latest['location'] == country]['death_rate'].values[0]
    print(f"{country}: Death Rate = {rate:.2f}%")
```

```python
##VISUALIZE BAR CHARTS
# Get latest data per country
latest_data = df.sort_values('date').groupby('location', as_index=False).last()

# Top 10 countries by total cases
top_countries = latest_data.nlargest(10, 'total_cases')

# Bar chart
plt.figure(figsize=(10, 6))
plt.barh(top_countries['location'], top_countries['total_cases'], color='skyblue')
plt.title("Top 10 Countries by Total COVID-19 Cases")
plt.xlabel("Total Cases")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
###heatmaapp

# Choose numeric columns for correlation
corr_columns = ['total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'total_vaccinations']
corr_data = df[corr_columns].dropna()

# Compute correlation matrix
corr_matrix = corr_data.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between COVID-19 Metrics")
plt.tight_layout()
plt.show()
```
